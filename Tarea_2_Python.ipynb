```python
import pandas as pd
inmuebles = pd.read_csv ("/content/drive/MyDrive/Startup/inmuebles_bogota.csv")
```

```python
columnas = {"Baños":"Banos", "Área":"Area"}
inmuebles = inmuebles.rename(columns=columnas)
```

```python
informacion_valor = inmuebles.Valor.str.split(expand=True)
inmuebles ["Moneda"] = informacion_valor [0]
inmuebles ["Precio"] = informacion_valor [1]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,10))
grafica = sns.histplot (x="informacion_valor",
                        hue = "Tipo",
                        data=inmuebles.loc[inmuebles.Tipo.isin(['Apartamento', 'Finca', 'Lote'])],
                        kde=True
                        )
grafica.set_title('Área por tipo de inmueble *Apartamento, finca y lote*')
plt.xlim(0,1000)
plt.ylim(0,1000)
plt.xlabel("Precio")
plt.ylabel("Cantidad inmuebles",fontsize=12)
plt.show()
```
